# Prepare Data

```python
# Convert string representation of list to actual list
toun_4_inversion_90_30['Best_fitness'] = toun_4_inversion_90_30['Best_fitness'].apply(ast.literal_eval)
toun_4_inversion_90_10['Best_fitness'] = toun_4_inversion_90_10['Best_fitness'].apply(ast.literal_eval)
toun_4_arithmetic_90_30['Best_fitness'] = toun_4_arithmetic_90_30['Best_fitness'].apply(ast.literal_eval)
toun_4_arithmetic_90_10['Best_fitness'] = toun_4_arithmetic_90_10['Best_fitness'].apply(ast.literal_eval)
toun_15_inversion_90_30['Best_fitness'] = toun_15_inversion_90_30['Best_fitness'].apply(ast.literal_eval)
toun_15_inversion_90_10['Best_fitness'] = toun_15_inversion_90_10['Best_fitness'].apply(ast.literal_eval)
#toun_15_arithmetic_90_30['Best_fitness'] = toun_15_arithmetic_90_30['Best_fitness'].apply(ast.literal_eval)
#toun_15_arithmetic_90_10['Best_fitness'] = toun_15_arithmetic_90_10['Best_fitness'].apply(ast.literal_eval)
```

```python
### Box plot do each Generation in each Experiment
```

```python
# Convert the 'Best_fitness' column of each data frame to a list of lists
fitness_data1 = toun_4_inversion_90_30['Best_fitness'].tolist()
fitness_data2 = toun_4_inversion_90_10['Best_fitness'].tolist()
```

```python
# Create a dictionary to store the fitness values for each index
fitness_values1 = {}
# Iterate over each row and extract fitness values for each index
for row in fitness_data1:
    for i, value in enumerate(row):
        if i not in fitness_values1:
            fitness_values1[i] = []
        fitness_values1[i].append(value)
# Create a list to store the data for each index
data_toun_4_inversion_90_30 = [fitness_values1[i] for i in range(len(fitness_values1))]
```

```python
# Create a dictionary to store the fitness values for each index
fitness_values2 = {}
# Iterate over each row and extract fitness values for each index
for row in fitness_data2:
    for i, value in enumerate(row):
        if i not in fitness_values2:
            fitness_values2[i] = []
        fitness_values2[i].append(value)
# Create a list to store the data for each index
data_toun_4_inversion_90_10 = [fitness_values2[i] for i in range(len(fitness_values2))]
```

```python
# Create a box plot for each index
bp1 = plt.boxplot(data_toun_4_inversion_90_30, patch_artist=True)
bp2 = plt.boxplot(data_toun_4_inversion_90_10, patch_artist=True)

# Set colors
colors = ['blue', 'green']  # define your colors list
for bplot, color in zip((bp1, bp2), colors):
    for patch in bplot['boxes']:
        patch.set_facecolor(color)

# Add labels to the x-axis
plt.xticks(range(1, len(fitness_values1) + 1), range(len(fitness_values1)))
plt.xticks(range(1, len(fitness_values2) + 1), range(len(fitness_values2)))

# Add a title and labels to the plot
plt.title('Box Plot of Best Fitness by Generation, Investion Mutation')
plt.xlabel('Generation')
plt.ylabel('Fitness')

# Create custom artists for legend
legend_artists = [plt.Line2D([0], [0], color=color, marker='o', linestyle='', markersize=10) for color in colors]
plt.legend(legend_artists, ['Mutation Prob 30%', 'Mutation Prob 10%'], loc='best')  # Adjust the labels according to your datasets

# Display the plot
plt.show()
```
